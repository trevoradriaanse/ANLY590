```python
def display_activation(f, every=False):
    I = np.linspace(-2, 2)
    if not every:
        plt.plot(I, f(I))
    else:
        plt.plot(I, sigmoid(I), label='sigmoid')
        plt.plot(I, tanh(I), label='tanh')
        plt.plot(I, relu(I), label='relu')
        plt.plot(I, leaky_relu(I), label='leaky relu')
        plt.legend()
    plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
def relu(x):
    return np.maximum(0, x)
def leaky_relu(x, a=0.5):
    return np.maximum(0, a*x)
```
